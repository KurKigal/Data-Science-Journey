```python
df['Units Sold_Filled'] = df['Units Sold'].fillna(df['Units Sold'].mean())
```

```python
## Renaming Columns
df.rename(columns={'Units Sold': 'Units_Sold', 'Order Date': 'Order_Date'}, inplace=True)
```

```python
df.head()
```

```python
df['Value_new'] = df['Units_Sold'].astype(float)
df.head()
```

```python
df.dtypes 
```

```python
df['Value_new'] = df['Units_Sold'].fillna(df['Units_Sold'].mean()).astype(int)
df.head()
```

```python
df['Doubled_New_Value'] = df['Value_new'].apply(lambda x: x * 2)
df.head()
```

```python
## Data Aggregating and Grouping
grouped_mean = df.groupby('Product Category')['Unit Price'].mean()
print(grouped_mean)
```

```python
grouped_sum = df.groupby(['Product Category','Region'])['Unit Price'].sum()
print(grouped_sum)
```

```python
## Aggregate Multiple Functions
grouped_agg = df.groupby('Region')['Unit Price'].agg(['mean', 'sum', 'count'])
grouped_agg
```

```python
## Merging and Joining Dataframes
df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4, 5, 6]})
```

```python
## Merge DataFrames on the Key Columns
merged = pd.merge(df1, df2, on='Key', how='inner')
merged
```

```python
merged = pd.merge(df1, df2, on='Key', how='outer')
merged
```

```python
merged = pd.merge(df1, df2, on='Key', how='left')
merged
```

```python
merged = pd.merge(df1, df2, on='Key', how='right')
merged
```
