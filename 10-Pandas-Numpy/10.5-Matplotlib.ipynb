```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

## Create a line plot

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

## Create a customized line plot
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Line Plot Example')
plt.grid(True)
plt.show()
```

```python
## Multiple Plots
## Sample Data
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [2,3,5,7,11]

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1) ##subplot(nrows, ncols, index, **kwargs)
plt.plot(x, y1, color='blue', label='y1')
plt.title('Plot 1')
plt.legend()

plt.subplot(1, 2, 2) 
plt.plot(x, y2, color='green', label='y2')
plt.title('Plot 2')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
## Bar Plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 4, 6]

plt.bar(categories, values, color='orange')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.show()
```

```python
## Histograms
# Histograms are used to represent the distribution of numerical data.

## Create a histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
plt.hist(data, bins=5, color='purple', edgecolor='black', alpha=0.7) # alpha is for transparency
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

```python
## Create a Scatter Plot
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.scatter(x, y, color='red', marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.grid(True)
plt.show()
```

```python
## Pipe Chart

labels = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 4, 6]

plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140) 
## Startangle is the angle at which the pie chart starts
## Autopct is used to display the percentage on the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart Example')
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [30, 20, 40, 10, 50]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink']
explode = (0.2, 0, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart Example')
plt.show()
```

```python
## Sales Data Visualization

import pandas as pd

data = pd.read_csv('Online_Sales_Data.csv')
data.head()
```

```python
## Plot total sales by product category
category_sales = data.groupby('Product Category')['Total Revenue'].sum()
category_sales.plot(kind='bar')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
## Plot Sales Trend Over Time
sales_trend = data.groupby('Date')['Total Revenue'].sum()
sales_trend.plot(kind='line')
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.show()
```
